```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

```

```python
data = pd.read_csv('train.csv')

data.head()
```

```python
data.isna().sum()
```

```python
test_data = pd.read_csv('test.csv')
test_ids = test_data['id']
test_data.head()
```

```python
test_data.drop(['state','area_code'],axis=1,inplace=True)
```

```python
test_data.head()
```

```python
le  = LabelEncoder()
cat_cols = ['international_plan','voice_mail_plan','churn']
for col in cat_cols:
  data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
cat_col = ['international_plan','voice_mail_plan']
for i in cat_col:
  test_data[i] = le.transform(test_data[i])
test_data.head()
```

```python
test_data.isna().sum()
```

```python
y = data['churn']
X = data.drop(['churn'],axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=10)
```

```python
log_reg = LogisticRegression()
rdm_model = RandomForestClassifier()
dt_model = DecisionTreeClassifier()
xg_model = XGBClassifier()
```

```python
log_reg.fit(X_train,y_train)
rdm_model.fit(X_train,y_train)
dt_model.fit(X_train,y_train)
xg_model.fit(X_train,y_train)
```

```python
y1_pred = log_reg.predict(X_test)
y2_pred = rdm_model.predict(X_test)
y3_pred = dt_model.predict(X_test)
y4_pred = xg_model.predict(X_test)
```

```python
accuracy_score(y_test,y1_pred)
```

```python
accuracy_score(y_test,y2_pred)
```

```python
accuracy_score(y_test,y3_pred)
```

```python
accuracy_score(y_test,y4_pred)
```

```python
test_data.drop(['id'],axis=1,inplace=True)
test_data.head()
```

```python
submission_preds = xg_model.predict(test_data)
```

```python
df = pd.DataFrame({
    'id' : test_ids.values,
    'churn': submission_preds
})
```

```python
df.head()
```

```python
df.replace((1,0),('yes','no'),inplace=True
           )
```
