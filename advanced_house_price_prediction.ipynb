```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('train.csv')
```

```python
drop_cols = ['Alley','PoolQC','Fence','MiscFeature','FireplaceQu']
data.drop(drop_cols,axis=1,inplace=True)
```

```python
data.select_dtypes(include='object').isnull().sum()
```

```python
# selecting all the integer columns from dataset
int_features = [i for i in data.columns if data.dtypes[i]=='int']
```

```python
int_features
```

```python
# selecting all the categorical columns
cat_features = [i for i in data.columns if data.dtypes[i]=='object']
cat_features
```

```python
float_features = [i for i in data.columns if data.dtypes[i]=='float']
float_features
```

```python
len(cat_features) + len(int_features) + len(float_features)
```

```python
for col in float_features:

  data[col].fillna(data[col].mean(),inplace=True)

```

```python
data.select_dtypes(include='float').isnull().sum()
```

```python
for col in cat_features:
  if data[col].isna().sum() != 0:
    mode = str(data[col].mode())
    data[col].fillna(mode,inplace=True)
data.select_dtypes(include='object').isnull().sum()
```

```python
le = LabelEncoder()
```

```python
for col in cat_features:
  data[col] = le.fit_transform(data[col])
  print(le.classes_)
```

```python
data.head()
```

```python
y = data['SalePrice']
X = data.drop(['SalePrice','Id'],axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,
                                                 test_size=0.2,
                                                 random_state=10)
```

```python
Rf_model = RandomForestRegressor()
Xg_model = XGBRegressor()
Dt_model = DecisionTreeRegressor()
```

```python
Rf_model.fit(X_train,y_train)
Xg_model.fit(X_train,y_train)
Dt_model.fit(X_train,y_train)
```

```python
Rf_pred = Rf_model.predict(X_test)
Xg_pred = Xg_model.predict(X_test)
Dt_pred = Dt_model.predict(X_test)
```

```python
mean_absolute_error(Rf_pred,y_test)
```

```python
mean_absolute_error(Xg_pred,y_test)
```

```python
mean_absolute_error(Dt_pred,y_test)
```
